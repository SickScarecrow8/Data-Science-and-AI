```python
dimension = int(input("Escribe la dimensión de la matriz: "))
matriz = []

for i in range(dimension):
    for j in range(dimension):
        num = int(input(f"Escribe el número de la posición ({i + 1}, {j + 1}): "))
        matriz.append(num)

matriz = np.array(matriz).reshape(dimension, dimension)
print(matriz)

detMatriz = np.linalg.det(matriz)
print(f"El determinante de la matriz: {detMatriz}")
```

```python
def calcular_focos():
    modelos = np.array(["M1", "M2", "M3", "M4"])
    defectos = np.array([0.02, 0.05, 0.08, 0.10])

    # Matrices para almacenar la cantidad de focos por tipo y modelo
    focos = {
        "T": np.zeros(4, dtype=int),
        "O": np.zeros(4, dtype=int)
    }

    for tipo in focos:
        print(f"Ingrese la cantidad de focos {tipo} por modelo:")
        focos[tipo] = np.array([int(input(f"{modelo}: ")) for modelo in modelos])

    # Cálculos
    transparentes_buenos = np.sum(focos["T"] * (1 - defectos))
    opacos_defectuosos = np.sum(focos["O"] * defectos)

    # Determinar el modelo con más defectuosos
    defectuosos_totales = {
        "T": focos["T"] * defectos,
        "O": focos["O"] * defectos
    }

    max_tipo, max_modelo_idx = max(
        [(tipo, idx) for tipo in defectuosos_totales for idx in range(4)],
        key=lambda x: defectuosos_totales[x[0]][x[1]]
    )
    max_modelo = modelos[max_modelo_idx]
    max_defectos = defectuosos_totales[max_tipo][max_modelo_idx]

    # Resultados
    print(f"Cantidad de focos transparentes en buen estado: {int(transparentes_buenos)}")
    print(f"Cantidad de focos opacos defectuosos: {int(opacos_defectuosos)}")
    print(f"El tipo y modelo de foco que más se produce en mal estado es {max_tipo}-{max_modelo} con {int(max_defectos)} defectuosos.")

# Ejecutar la función
calcular_focos()
```
