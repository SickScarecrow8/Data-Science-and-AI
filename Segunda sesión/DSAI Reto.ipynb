**Recuperar la información del Archivo employee_attrition.csv**

```
# Recuperamos los datos del archivo csv y creamos un dataframe llamado "empleados"
empleados = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data Science and AI/Segunda sesión/Employee_Attrition.csv")
empleados
```

```
empleados.info()
```

**El tiempo de servicio promedio de los trabajadores**

```
# Utilizamos la función mean para obtener el promedio de toda la columna "Time_of_service"
promedio_tiempo_servicio = empleados['Time_of_service'].mean()
print(f"El tiempo de servicio promedio de todos los empleados es de: {promedio_tiempo_servicio}")
```

**El porcentaje de hombres y mujeres en el archivo**

```
# Agrupamos la columna "Gender" en datos de tipo Female y Male y dividimos la cuenta entre el número total de empleados y multiplicamos por 100
# para obtener el promedio
porcentaje_f_m = empleados['Gender'].groupby(empleados['Gender']).count() / empleados['Gender'].count() * 100
porcentaje_f_m
```

```
# Usando la función value_counts con el parámetro normalize como True y multiplicando por 100 llegamos a un resultado más corto
empleados['Gender'].value_counts(normalize = True) * 100
```

**El tiempo máximo sin promoción de las personas que llevan más de 15 años de servicio**

```
# Se selecciona a los empleados con más de 15 años de servicio y se obtiene el valor máximo usando la función max
maxtime_sincepromotion = empleados[empleados['Time_of_service'] > 15]['Time_since_promotion'].max()
print(f"El tiempo máximo sin promoción que llevan los empleados con más de 15 años de servicio es de: {maxtime_sincepromotion}")
```

**La edad promedio en la que ingresaron las personas a la empresa, genera una nueva columna derivada mediante la formula Age - Time_of_service**

```
empleados['Average_age'] = empleados['Age'] - empleados['Time_of_service']
empleados['Average_age']
```

```
# La edad promedio de ingreso de los empleados
empleados['Average_age'].mean()
```

**Un dataframe con las columnas Employee_ID, Relationship, Decision_skill_process que sólo contenga los registros de mujeres, solteras y con compensación tipo 2 o 3**

```
# Se utiliza la función loc para obtener los registros de los empleados que son mujeres y que su status de relación sea single
# También se ocupa la funcion isin para encontrar su tipo de compensación
femsingletype23 = empleados[['Employee_ID', 'Relationship_Status', 'Decision_skill_possess']].loc[((empleados['Gender'] == "F")
& (empleados['Relationship_Status'] == "Single")
& (empleados['Compensation_and_Benefits'].isin(["type2", "type3"])))]
femsingletype23
```

**Una columna derivada con el valor "on time" si cuando tuvo su última promoción tenía menos de 30 años. En caso contrario colocar el valor de "out of time"**

```
# Se crea la columna derivada a través de una función lambda en donde si la diferencia de la edad menos el tiempo desde la promoción es menor a 30 se escribe "on time"
# en caso de ser mayor a 30 se escribe "out of time"
empleados['Less_than_30'] = (empleados['Age'] - empleados['Time_since_promotion']).apply(lambda x: "on time" if x < 30 else "out of time")
empleados['Less_than_30']
```

```
# Se usa la función where de numpy para llegar al mismo resultado anterior, ahorrando código
empleados['Less_than_30'] = np.where((empleados['Age'] - empleados['Time_since_promotion']) < 30, "on time", "out of time")
empleados['Less_than_30']
```

**Guardar los dataframes resultantes**

```
# Convertimos a archivos xlsx los dataframes que obtuvimos
pd.DataFrame([promedio_tiempo_servicio], columns=['promedio_tiempo_servicio']).to_excel("promedio_tiempo_servicio.xlsx", index = False)
pd.DataFrame(porcentaje_f_m).to_excel("porcentaje_f_m.xlsx")
pd.DataFrame([maxtime_sincepromotion], columns=['maxtime_sincepromotion']).to_excel("maxtime_sincepromotion.xlsx", index = False)
pd.DataFrame(empleados['Average_age']).to_excel("empleados_average_age.xlsx")
pd.DataFrame(femsingletype23).to_excel("femsingletype23.xlsx")
pd.DataFrame(empleados['Less_than_30']).to_excel("empleados_less_than_30.xlsx")
pd.DataFrame(empleados).to_excel("empleados.xlsx")
```
