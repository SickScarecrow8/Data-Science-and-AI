```
retweetsdf = tweets.loc[tweets['retweets'] > 0, ['user_name', 'retweets']]
retweetsdf['retweets'].mean()
```

```
favorites = tweets.loc[tweets['favorites'] > 100, ['user_name', 'favorites']]
favorites.nlargest(5, 'favorites')
```
