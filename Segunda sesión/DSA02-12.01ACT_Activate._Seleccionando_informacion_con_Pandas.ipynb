5. Agregar una nueva columna (size) con el valor LARGE cuando la longitud (length) sea mayor a 2000 y el valor SMALL en caso contrario. Además de la nueva columna solamente debe contener las columnas name, passengers, cargo_capacity y starship_class.

```
# Convert the 'length' column to numeric, handling errors by coercing them to NaN
starships["length"] = pd.to_numeric(starships["length"], errors = "coerce")

# Now apply the lambda function
starships["size"] = starships["length"].apply(lambda x: "LARGE" if x > 2000 else "SMALL")
starships[["name", "passengers", "cargo_capacity", "starship_class", "size"]]
```
