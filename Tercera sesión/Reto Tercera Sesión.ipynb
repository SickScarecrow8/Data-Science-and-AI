```python
# Generamos DataFrames con cada archivo csv

train1 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Data Science and AI/Tercera sesión/Train.csv')
train2 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Data Science and AI/Tercera sesión/Train2.csv')
test1 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Data Science and AI/Tercera sesión/Test.csv')
test2 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Data Science and AI/Tercera sesión/Test2.csv')
```

```python
# Concatenamos los dataframes que contengan la misma estructura usando la función concat()

merge1 = pd.concat([train1, test1], axis = 0)
merge2 = pd.concat([train2, test2], axis = 0)
```

```python
merge1
```

```python
merge2
```

```python
# Usando la función merge y tomando como índice la columna de "Employee_ID" unimos los dataframes resultantes de la función anterior

final_merge = pd.merge(merge1, merge2, on = 'Employee_ID')
final_merge
```

```python
final_merge.info()
```

```python
# Notamos que los nombres de las coolumnas tienen un espacio al principio, entonces usamos la función str.strip() aplicada a las columnas

final_merge.columns = final_merge.columns.str.strip()
final_merge.info()
```

**a: El porcentaje de personas que usan Windows/Mac/Linux/No registrado como su sistema operativo en su computadora**

```python
# Usamos la función value_counts() con el parámetro dropna = False para que cuente los sistemas operativos no registrados

avg_computerOS = final_merge['Computer_OS'].value_counts(dropna = False, normalize = True) * 100
avg_computerOS
```

**b: El porcentaje de personas que usan iOS/Android/No registrado como su sistema operativo en su dispositivo móvil**

```python
# Usamos la función value_counts() con el parámetro dropna = False para que cuente los sistemas operativos no registrados

avg_mobileOS = final_merge['Mobile_OS'].value_counts(dropna = False, normalize = True) * 100
avg_mobileOS
```

**c: La edad promedio de la gente que utiliza cada sistema operativo de escritorio**

```python
# Agrupamos los datos por sistema operativo y a la columna de edad le aplicamos la función mean()

avg_age_computerOS = final_merge.groupby('Computer_OS')['Age'].mean()
avg_age_computerOS
```

**d: El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio**

```python
# Agrupamos los datos por sistema operativo y a la columna de nivel educativo le aplicamos la función mean()

avg_educationLevel = final_merge.groupby('Computer_OS')['Education_Level'].mean()
avg_educationLevel
```

**e: El número total de tickets generados por la gente con cada sistema operativo de escritorio**

```python
# Agrupamos los datos en sistema operativo y en cantidad de tickets y utilizamos la función value_counts() con el parámetro dropna = False,
# para que cuente los datos nulos y el parámetro sort = False para que preserve el orden de los datos

avg_computerTickets = final_merge.groupby('Computer_OS')['Computer_tickets'].size()
avg_computerTickets
```

**Enviar el resultado de cada solicitud de información a un archivo XLS[X], donde el resultado de cada consulta debe ir en su propia hoja de trabajo**

```python
!pip install openpyxl xlsxwriter

with pd.ExcelWriter('consultas.xlsx', engine = 'xlsxwriter') as writer:

    avg_computerTickets.to_excel(writer, sheet_name = 'avg_computerTickets')
    avg_age_computerOS.to_excel(writer, sheet_name = 'avg_age_computerOS')
    avg_computerOS.to_excel(writer, sheet_name = 'avg_computerOS')
    avg_educationLevel.to_excel(writer, sheet_name = 'avg_educationLevel')
    avg_mobileOS.to_excel(writer, sheet_name = 'avg_mobileOS')
```
