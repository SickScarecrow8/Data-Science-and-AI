1. El primer paso es carga el archivo en un DataFrame

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
dataframe = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Data Science and AI/Tercera sesión/Ejemplo1.xls')
```

```python
dataframe
```

2. El siguente paso es eliminar todos los registros que sean de Estados Unidos

```python
dataframe = dataframe[dataframe['Country'] != 'United States']
dataframe
```

3. Luego, Muestra el número de personas registradas de cada país (ya sin los de Estados Unidos). Se espera una tabla que diga país y número de personas.

```python
dataframe['Country'].groupby(dataframe['Country']).count().rename(index = {'Country': 'País'}).rename("Número de personas")
```

4. Ahora, actualiza el nombre de los países para que el nombre esté en español

```python
dataframe.loc[(dataframe['Country'] == "France"), 'Country'] = "Francia"
dataframe.loc[(dataframe['Country'] == "Great Britain"), 'Country'] = "Gran Bretaña"
dataframe
```

5. el siguiente paso es cambiar el id de <b>Judie Claywell Female France</b> de 7569 a 99999

```python
dataframe.loc[(dataframe['First Name'] == "Judie") & (dataframe['Last Name'] == "Claywell"), "Id"] = 99999
dataframe
```

6. Ahora, crea un archivo de Excel en tu drive, en la misma carpeta en la que estas trabajando, que contenga exactamente el mismo primer renglón del archivo descargado(el de excel), pero ahora Inserta 3 registros con gente de México con ids de 5 dígitos que no inicien en cero

```python
dataframe2 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Data Science and AI/Tercera sesión/Ejemplo3.xlsx')
dataframe2
```

7. Carga este segundo archivo como un dataFrame y concaténalo a los datos del archivo descargado

```python
dataframe = pd.concat([dataframe, dataframe2])
dataframe
```

8. Vuelve a mostrar el número de personas registradas de cada país

```python
gentePorPais = dataframe['Country'].groupby(dataframe['Country']).count().rename(index = {'Country': 'País'}).rename("Número de personas")
gentePorPais
```
