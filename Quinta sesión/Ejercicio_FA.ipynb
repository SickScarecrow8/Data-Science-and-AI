```python
# Librerías para tratamiento de datos
import numpy as np
import pandas as pd
import statsmodels.api as sm
# Procesado y modelado
from factor_analyzer import FactorAnalyzer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
# Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sm.datasets.get_rdataset('iris', 'datasets')
datos = iris.data
datos
```

```python
print('Media de cada variable')
datos.select_dtypes(include = np.number).mean()
```

```python
print('Desviación estándar de cada variable')
datos.select_dtypes(include = np.number).std()
```

```python
escalar = StandardScaler()
```

```python
# Iris
features = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
# Separating out the features
x = datos.loc[:, features]
# Separating out the target
y = datos.loc[:, ['Species']]
```

```python
datos_escalados = escalar.fit_transform(x)
datos_escalados
```

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value, p_value = calculate_bartlett_sphericity(datos_escalados)
chi_square_value, p_value
```

```python
fa = FactorAnalyzer()
fa.fit(datos_escalados)
ev, v = fa.get_eigenvalues()
ev
```

```python
fa = FactorAnalyzer(n_factors = 1)
fa.fit(datos_escalados)
```

```python
pd.DataFrame(fa.loadings_)
```

```python
pd.DataFrame(
    data = fa.loadings_,
    columns = ['Factor 1'],
    index = x.columns
)
```

```python
fa2 = FactorAnalyzer(n_factors = 2)
fa2.fit(datos_escalados)
```

```python
pd.DataFrame(fa2.loadings_)
```

```python
pd.DataFrame(
    data = fa2.loadings_,
    columns = ['Factor 1', 'Factor 2'],
    index = x.columns
)
```

```python
fa.get_factor_variance()
```

```python
fa2.get_factor_variance()
```

```python
datos_fa = pd.DataFrame(fa2.transform(datos_escalados))
datos_fa
```

```python
conjunto_fa = pd.concat([datos_fa, datos[['Species']]], axis = 1)
conjunto_fa
```
