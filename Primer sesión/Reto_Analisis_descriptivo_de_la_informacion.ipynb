# Abrir el archivo e iterar sobre cada renglón

# Edad promedio de los empleados

```python
file_handler = open('/content/employees.csv')
next(file_handler)

edad = []

for line in file_handler:
    datos = line.split(',')
    if datos[2].isdigit():
        edad.append(int(datos[2]))

print(f"La edad promedio de los empleados es: {sum(edad) / len(edad)}")

file_handler.close()

```

# Recuperar género de los empleados

```python
file_handler = open('/content/employees.csv')
next(file_handler)

empleados_hombres = []
empleados_mujeres = []

for line in file_handler:
    datos = line.split(',')

    if datos[1] == 'M':
        empleados_hombres.append((datos[1]))
    elif datos[1] == 'F':
        empleados_mujeres.append((datos[1]))

print(empleados_hombres)
print(empleados_mujeres)

print()

print(f"La cantidad de empleados hombres es: {len(empleados_hombres)}")
print(f"La cantidad de empleados mujeres es: {len(empleados_mujeres)}")
print("Hay más empleados mujeres.")

file_handler.close()
```
