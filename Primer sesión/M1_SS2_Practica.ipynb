***Primera opción***

```python
file_handler = open('/content/planets.csv')
next(file_handler)

min_nom = ''
min_rot = 999999
max_nom = ''
max_orb = 0

for line in file_handler:
    columnas = line.split(",")
    if columnas[0] != "NA" and columnas[1] != "NA" and columnas[2] != "NA":
        nombre = columnas[0]
        rotacional = int(columnas[1])
        orbital = int(columnas[2])

        if rotacional < min_rot:
            min_rot = rotacional
            min_nom = nombre
        elif rotacional == min_rot:
            min_nom += ', ' + nombre

        if orbital > max_orb:
            max_orb = orbital
            max_nom = nombre
        elif orbital == max_orb:
            max_nom += ', ' + nombre

print('El planeta con menor periodo rotacional es', min_nom, min_rot)
print('El planeta con mayor periodo orbital es', max_nom, max_orb)

file_handler.close()
```

***Segunda opción***

```python
file_handler = open('/content/planets.csv')
next(file_handler)

nombres = []
rotacionales = []
orbitales = []

for line in file_handler:
    columnas = line.split(",")
    if columnas[0] != "NA" and columnas[1] != "NA" and columnas[2] != "NA":
        nom = columnas[0]
        rot = int(columnas[1])
        orb = int(columnas[2])
        nombres.append(nom)
        rotacionales.append(rot)
        orbitales.append(orb)

file_handler.close()

min_rot = min(rotacionales)
ind_min = rotacionales.index(min_rot)
max_orb = max(orbitales)
ind_max = orbitales.index(max_orb)

print('El planeta con menor periodo rotacional es', nombres[ind_min], min_rot)
print('El planeta con mayor periodo orbital es', nombres[ind_max], max_orb)
```

# Imprimir todos los de menor periodo rotacional
