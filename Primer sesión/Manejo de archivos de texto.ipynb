***Ejemplo***

Contar el número de renglones que contiene un archivo

```
file_handler = open('/content/sample_data/prueba.txt', 'r')
count = 0
for line in file_handler:
  count = count + 1
print('Total number of lines:', count)
file_handler.close()
```

***Ejercicios***

1. Calcular la suma de los números que se encuentran en el archivo. El resultado debe ser igual a 130.









```
file_handler = open('/content/sample_data/prueba.txt', 'r')

sum = 0

for line in file_handler:
    sum = sum + int(line)
print(f"La suma total de las líneas es de {sum}.")

file_handler.close()
```

2. Calcular el promedio de los números que se encuentran en el archivo. El resultado debe ser igual a 26.



```
file_handler = open('/content/sample_data/prueba.txt', 'r')

sum = 0
count = 0

for line in file_handler:
    count = count + 1
    sum = sum + int(line)

avg = sum / count

print(f"El primedio total de los números en la lista es de {avg}")
file_handler.close()
```

3. Obtener el número más grande y el número más pequeño en el archivo. Los resultados deben ser 81 y 3.  

```
file_handler = open('/content/sample_data/prueba.txt', 'r')

max = -100
min = 100

for line in file_handler:
    if int(line) > max:
        max = int(line)
    if int(line) < min:
        min = int(line)

print(f"El número más grande es {max} y el número más pequeño es {min}")
file_handler.close()
```
