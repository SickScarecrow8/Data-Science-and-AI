***Ejemplo***

Calcular el área de un triángulo.
<br>Entradas: base y altura
<br>Salida: área del triángulo

```
def triangle_area(b, h):
  return b * h / 2

a = triangle_area(2.0, 5.0)
print('Area:', a)
```

1. Calcular la sumatoria de un número.







```
def sumatoria_num(num):

    sum = 0
    for i in range(1, (num + 1)):
        sum = sum + i
    return sum

print(sumatoria_num(10))
```

2. Identificar si un año es bisiesto o no, considerando las siguientes dos reglas:

*   Un año es bisiesto si es divisible entre 4.
*   Un año es bisiesto si es divisible entre 100, pero también es divisible entre 400.



```
def bisiesto(year):

    if (year % 4 == 0) or ((year % 100 == 0) and (year % 400 == 0)):
        print(f"El año {year} es año bisiesto.")
    else:
        print(f"El año {year} no es año bisiesto.")

bisiesto(2024)
```

3. Calcular la edad de un perro en años humanos, teniendo en cuenta que el primer año de un perro equivale a 15 años humanos, el segundo año de un perro equivale a 9 años humanos y de ahí en adelante cada año equivale a 4 años humanos.

*   Un perro de 1 año tendría 15 años humanos  
*   Un perro de 2 años tendría 24 años humanos
*   Un perro de 3 años tendría 28 años humanos

```
def edadperro(edad):

    if edad == 1:
        print("El perro tiene 15 años humanos.")

    elif edad > 1 and edad < 3:
        print("El perro tiene 24 años humanos.")

    elif edad >= 3:
        edad = ((edad - 2) * 4) + 24
        print(f"El perro tiene {edad} años humanos.")

    elif edad < 1 and edad > 0:
        print("El perro aún no alcanza una edad suficiente para calcularla.")

    else:
        print("Verifica la edad del perro.")

edadperro(18)

```
