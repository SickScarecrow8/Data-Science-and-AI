# **TEMA 3. Estatutos de repetici√≥n**

##3.1 While

```
# Estatuto while simple
num = int(input('Dame un entero positivo: '))
i = 0

while i <= num:
    print(i)
    i = i + 1   # i += 1

print('Primera l√≠nea fuera del while')
```

```
# Estatuto break
calificacion = 0
promedio = 0
i = 0
while i < 10:
    calificacion_txt = input('Dame una calificaci√≥n o un n√∫mero negativo o mayor a 10 para terminar')
    calificacion = float(calificacion_txt)
    print(calificacion)
    if calificacion < 0 or calificacion > 10:
        break
    promedio = promedio + calificacion
    i = i + 1

print('Primera l√≠nea fuera del while')
promedio = promedio / i
print('El promedio es:', promedio)
```

```
# Estatuto continue
i = 0
promedio = 0
while i < 5:
    grade_txt = input('Dame una calificaci√≥n o un n√∫mero negativo para un NP')
    grade = float(grade_txt)
    print(grade)
    if grade < 0:
        continue
    promedio = promedio + grade
    i = i + 1

print('Primera l√≠nea fuera del while')
promedio = promedio / i
print('El promedio es:', promedio)
```

##3.2 For

```
# Estatuto for iterando sobre una cadena de caracteres
txt = input('Dame un texto: ')

for letra in txt:
    print(letra)
```

```
# Estatuto for iterando sobre una cadena de caracteres
# Contar caracteres
txt = input('Dame un texto: ')
i = 0
for caracter in txt:
    print(caracter)
    i = i + 1
print('N√∫mero de caracteres:', i)
```

```
# Estatuto for iterando sobre una lista
lst = [10, 74, 23, 80, 9, 30]
avg = 0
for num in lst:
    print(num)
    avg = avg + num   # avg += num
avg = avg / len(lst)

print(f'N√∫mero de elementos: {len(lst)}')
print(f'Promedio: \033[1;34m{avg:.2f}\033[0m')
```

# **TEMA 4. Definici√≥n de Funciones**

##4.1 Estructura de una funci√≥n

```
# Definici√≥n de una funci√≥n simple
def greet():
    print("Hello!")

greet()
```

```
# Definici√≥n de una funci√≥n con un par√°metro
def greet(name):
    print('Hello', name)

greet('Juan Carlos')
greet('Patricia')
```

```
# Definici√≥n de una funci√≥n que tiene un par√°metro
# y un valor por defecto para el mismo
def greet(name = 'World'):
  print('Hello', name)

greet('Brenda')
greet()
```

```
# Definici√≥n de una funci√≥n que recibe dos par√°metros
def add(a, b):
    return a + b

x = add(2, 5)
print("Addition: ", x)
print("Addition: ", add(1.2, 7.5))
print("Addition: ", add(True, True))
print("Addition: ", add("a", "b"))
print("Addition: ", add(-1, 8))
```

```
def greet(times = 3):
    print("Hello " * times)

greet()
greet(5)
greet(8)
```

##4.2 Llamando a una funci√≥n

```
# Llamada a una funci√≥n con un par√°metro
def greet(name):
  print('Hello', name)

greet('Rosa')
```

```
# Definici√≥n de la funci√≥n
def factorial(n):
    fact = 1
    for x in range(1, n + 1):
        fact = fact * x
    return fact

# La funci√≥n combinatoria calcula el n√∫mero de posibles
# combinaciones al seleccionar un n√∫mero m de elementos
# de un conjunto de n elementos
def combinatory(n, m):
    return factorial(n) / (factorial(m) * factorial(n-m))

# Llamada a la funci√≥n
print(combinatory(6, 2))
print(combinatory(5, 3))
```

```
# Definici√≥n de una funci√≥n para calcular el √°rea de un tri√°ngulo

def triangle_area(b, h):
    return b * h / 2

area = triangle_area(2.0, 5.0)
print('√Årea:', area)
```

***Archivos vac√≠os: os stat y st_size***

```
import os

# Revisar si un archivo est√° vac√≠o
def file_is_not_empty(name):
  return os.stat(name).st_size > 0

file_name = '/content/sample_data/prueba.txt'

if file_is_not_empty(file_name):
  file_handler = open(file_name)
  for line in file_handler:
    print(line)
  file_handler.close()
else:
  print('The file is empty')
```

```
# Contar el n√∫mero de renglones de un archivo
file_handler = open('/content/sample_data/prueba.txt', 'r')
count = 0
for line in file_handler:
    count = count + 1
print('N√∫mero total de l√≠neas:', count)
file_handler.close()
```

***Recursivamente***

```
import glob

# Obtener todos los archivos con una misma extensi√≥n
# dentro de un directorio de forma recursiva
path = '/content/sample_data'
files = glob.glob(path + '/**/*.txt', recursive=True)
print(files)
```

***Iterador***

```
import glob

# Obtener todos los archivos con una misma extensi√≥n
# dentro de un directorio y guardarlos en un iterador
path = '/content/sample_data'
files = glob.iglob(path + '/*.txt')
for file in files:
  print(file)
```
