# **Análisis de los datos**
# Variables de correlación.
Muchas de las variables categóricas no muestran mucha correlación con respecto al attrition rate, sin embargo, ciertos parámetros de estas variables sí muestran una tendencia a incrementar éste índice. Por lo tanto lo más recomendable es dejar estas variables en el análisis, a excepción de unas cuántas que no aportan mucha información al estudio.
# Correcciones
1.   La variable attrition representa únicamente dos datos, por lo tanto lo recomendable es convertirla a valores binarios.
2.   Las variables DailyRate, HourlyRate y MonthlyRate representan los mismos datos, por lo tanto podemos únicamente conservar una de estas. Además esta variable no es del todo conocimiento del empleado y no debería afectar al attrition rate.
3.   La variable EmployeeCount tiene el mismo dato en todos los registros, por lo tanto no aporta nada al estudio, así como la variable Over18, entonces se eliminarán.
4.   La variable BusinessTravel representa una parte importante, pero es necesario tenerla en un formato numérico.
5.   La variable Gender también se puede tener en formato binario.
6.   La variable StandardHours solamente tiene el mismo dato en todos los registros, por lo tanto tampoco aporta nada al estudio.
7.   La variable DistanceFromHome podemos obtenerla en un rango numérico.
8.   La variable EmployeeNumber tiene todos sus registros diferentes, por lo tanto tampoco nos será de mucha ayuda.
# Valores nulos
El dataset no cuenta con valores nulos, por lo tanto no es necesario inputar valores en ninguna de las variables.

```python
data2 = data.drop(columns = ['HourlyRate', 'MonthlyRate', 'EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'])
data2
```

```python
data2['Attrition'] = data2['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0 if x == 'No' else x)
data2['BusinessTravel'] = data2['BusinessTravel'].apply(lambda x: 0 if x == 'Non-Travel' else 1 if x == 'Travel_Rarely' else 2 if x == 'Travel_Frequently' else x)
data2['OverTime'] = data2['OverTime'].apply(lambda x: 0 if x == 'No' else 1)
data2
```

```python
data3 = pd.get_dummies(data2, columns = ['Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus'], drop_first = True)
data3
```

```python
dataScaler = RobustScaler()
scaler = dataScaler.fit(data3)
dataScaled = scaler.transform(data3)
dataScaled
```

# **Creación de conjuntos de entrenamiento y prueba**

```python
train_df, test_df = train_test_split(datos, test_size = 0.2, random_state = 123)

X_train, y_train = train_df.drop(columns = ['Attrition'], axis = 1), train_df['Attrition']
X_test, y_test = test_df.drop(columns = ['Attrition'], axis = 1), test_df['Attrition']

print("Datos de entrenamiento: ", X_train.shape, y_train.shape)
print("Datos de prueba: ", X_test.shape, y_test.shape)
```

# **Modelos de aprendizaje**

# Bosque aleatorio

```python
rfcInicial_model = RandomForestClassifier(n_estimators = 10, max_depth = 3)
scores = pd.DataFrame(cross_validate(rfcInicial_model, X_train, y_train, cv = 10, return_train_score = True))
print("Score promedio de entrenamiento: ", scores['train_score'].mean())
print("Score promedio de validación: ", scores['test_score'].mean())
scores
```

```python
parameters = {'max_depth': [2, 3, 5, 7],
              'max_features': ['sqrt', 'log2', None],
              'n_estimators': [10, 30, 60, 100]}
rfc_grid = GridSearchCV(RandomForestClassifier(random_state = 1), param_grid = parameters,
                        return_train_score=True)
rfc_grid.fit(X_train, y_train)

print("Mejores hiperparámetros\n",rfc_grid.best_params_)
```

```python
rfc_model = rfc_grid.best_estimator_
rfc_model.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_estimator(rfc_model, X_test, y_test)
print(classification_report(y_test, rfc_model.predict(X_test)))
```

# Regresión Logística

```python
lrcInicial_model = LogisticRegression(C = 0.001, solver = 'newton-cg')
skf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 123)
scores = pd.DataFrame(cross_validate(lrcInicial_model, X_train, y_train, cv = skf, return_train_score = True))

print("Score promedio de entrenamiento: ", scores['train_score'].mean())
print("Score promedio de validación: ", scores['test_score'].mean())
scores
```

```python
parameters = {'C': np.logspace(-4, 4, 50),
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
lrc_rand = RandomizedSearchCV(LogisticRegression(random_state=1), n_iter=48,
                              param_distributions = parameters,
                              return_train_score=True)
lrc_rand.fit(X_train, y_train)

print("Mejores hiperparámetros\n", lrc_rand.best_params_)
```

```python
lrc_model = lrc_rand.best_estimator_
lrc_model.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_estimator(lrc_model, X_test, y_test)
print(classification_report(y_test, lrc_model.predict(X_test)))
```

# Bayes Ingenuo Gaussiano

```python
gnbc_model = GaussianNB()

scores = pd.DataFrame(cross_validate(gnbc_model, X_train, y_train, cv = 10, return_train_score = True))

print("Score promedio de entrenamiento: ", scores['train_score'].mean())
print("Score promedio de validación: ", scores['test_score'].mean())
scores
```

```python
gnbc_model.fit(X_train, y_train)

ConfusionMatrixDisplay.from_estimator(gnbc_model, X_test, y_test)
print(classification_report(y_test, gnbc_model.predict(X_test)))
```

# Máquina de vectores de soporte

```python
svmc_model = SVC(C = 0.01, kernel = "poly")

scores = pd.DataFrame(cross_validate(svmc_model, X_train, y_train, cv = 10, return_train_score = True))

print("Score promedio de entrenamiento: ", scores['train_score'].mean())
print("Score promedio de validación: ", scores['test_score'].mean())
scores
```

```python
parameters = {'C': [0.01, 0.1, 1, 10, 50],
              'kernel': ['linear', 'poly', 'rbf', 'sigmoid']}
svmc_grid = GridSearchCV(SVC(random_state = 1), param_grid = parameters,
                         return_train_score = True)
svmc_grid.fit(X_train, y_train)
print("Mejores hiperparámetros\n",svmc_grid.best_params_)
```

```python
svmc_model = svmc_grid.best_estimator_
svmc_model.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_estimator(svmc_model, X_test, y_test)
print(classification_report(y_test, svmc_model.predict(X_test)))
```

# Curva ROC

```python
plt.figure()
lw = 2
disp = RocCurveDisplay.from_estimator(rfc_model, X_test, y_test)
RocCurveDisplay.from_estimator(lrc_model, X_test, y_test, ax=disp.ax_)
RocCurveDisplay.from_estimator(gnbc_model, X_test, y_test, ax=disp.ax_)
RocCurveDisplay.from_estimator(svmc_model, X_test, y_test, ax=disp.ax_)
plt.plot([0, 1], [0, 1], color = "navy", lw=lw, linestyle = "--")
plt.title("ROC curve comparison")
plt.legend(loc = "lower right")
plt.show()
```
