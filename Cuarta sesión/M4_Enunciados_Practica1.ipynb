Del dataframe original obtén:

*   Box plot de la esperanza de vida (Healthy life expectancy) ¿Cómo se interpreta?

```
happiness.plot(kind = 'hist', y = 'Score', bins = 20)
```

```
happiness.plot(kind = 'box', y = 'Healthy life expectancy')
```

```
fig, axs = plt.subplots(2,3,figsize=(10,8))
plt.subplots_adjust(wspace=0.3)
columns = happiness.loc[:,'GDP per capita':'Perceptions of corruption'].columns
for column, ax in zip(columns, axs.ravel()):
  happiness.plot.scatter(x='Score', y=column, ax=ax)
```

```
sns.heatmap(happiness.loc[:,'Score':'Perceptions of corruption'].corr(), annot = True)
```

# **Pregunta 5**

Combina con el dataframe `countries` (world_table_country.csv) para graficar la felicidad promedio por continente

```
happiness_countries = happiness.merge(countries, left_on = 'Country or region', right_on = 'Name')
happiness_countries
```

```
happiness_countries.groupby('Continent')['Score'].mean()
```

```
happiness_countries.groupby('Continent')['Score'].mean().plot(kind = 'bar')
plt.show()
```
