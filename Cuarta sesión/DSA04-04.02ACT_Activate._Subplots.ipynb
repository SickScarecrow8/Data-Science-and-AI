**TEMA 1. Subplots**

1. Crea un dataframe `cities` con la información del archivo `world_table_city.csv`







```
cities = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Science and AI/Cuarta sesión/world_table_city.csv')
cities
```

2. A partir del dataframe `cities` obtén otro (`citiesMEX`) donde sólo almacenes las ciudades de México

```
citiesMEX = cities[cities['CountryCode'] == 'MEX']
citiesMEX
```

3. A partir del dataframe `cities` obtén otro (`citiesUSA`) donde sólo almacenes las ciudades de Estados Unidos

```
citiesUSA = cities[cities['CountryCode'] == 'USA']
citiesUSA
```

4. Agrupa la información de `citiesMEX` y `citiesUSA` para obtener la población total por distrito

```
districtUSA = citiesUSA[['District', 'Population']].groupby('District').sum()
```

```
districtMEX = citiesMEX[['District', 'Population']].groupby('District').sum()
```

5. Visualiza los resultados anteriores con gráficas de barras ubicadas en una matriz de 2x2

```
fig, axs = plt.subplots(2, 2, figsize = (15, 12))
axs[0, 0].bar(districtMEX.index, districtMEX['Population'])
axs[0, 0].tick_params(axis = 'x', rotation = 90)
axs[1, 1].bar(districtUSA.index, districtUSA['Population'])
axs[1, 1].tick_params(axis = 'x', rotation = 90)
plt.show()
```
