**TEMA 3. Gráficas para exploración de datos**

1. Lee en un nuevo dataframe la información del archivo `world_table_country.csv` para asegurarte que los cambios previos no afectarán los resultados de esta práctica

```
countries = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Science and AI/Cuarta sesión/world_table_country.csv')
countries
```

2. A partir de dataframe anterior obtén otro (`countriesNA`) donde sólo almacenes los países del continente norteamericano (North America) y empléalo para obtener los requerimientos de las siguientes preguntas

```
countriesNA = countries.loc[(countries['Continent'] == 'North America')]
countriesNA
```

3. Un histograma con el año de independencia (`IndepYear`), haciendo que tus clases tengan una separación de 50 años en el intervalo [1700 - 2000] ¿En qué período la mayor cantidad de países logró su independencia?

```
plt.hist(countriesNA['IndepYear'], bins = range(1700, 2000, 50))
plt.show()
```

4. Un diagrama de frecuencias para conocer la distribución de las formas de gobierno (`GovernmentForm`) ¿Cuál es la forma predominante en el continente?

```
sns.countplot(x = 'GovernmentForm', data = countriesNA)
plt.xticks(rotation = 90)
plt.show()
```

5. Un boxplot en el que se muestre la expectativa de vida (`LifeExpectancy`) por región (`Region`) ¿En qué región hay mayor dispersión de los datos y qué significado tiene?

```
countriesNA.boxplot(column = 'LifeExpectancy', by = 'Region', figsize = (10, 5), rot = 90)
plt.show()
```

6. Un scatterplot que muestre la relación entre la población (`Population`) y `GNP` ¿La correlación es positiva, negativa o nula?

```
import numpy as np
plt.scatter(x = np.log(countriesNA['Population']), y = np.log(countriesNA['GNP']))
plt.show()
```

```
countriesNA['GNP'].corr(countriesNA['Population'])
```

```
sns.pairplot(countriesNA)
plt.show()
```

7. Un heatmap que muestre el coeficiente de correlación entre todas las columnas del dataframe ¿Qué pares de variables o columnas están menos relacionadas para el conjunto de datos analizados?

```
sns.heatmap(round(countriesNA.select_dtypes(include = np.number).corr(), 2), annot = True, cmap = 'Blues')
plt.show()
```
