# **Pregunta 5**

Combina con el dataframe `countries` (world_table_country.csv) para graficar la felicidad promedio por continente

```
happiness_countries_merge = pd.merge(left=happiness,right=countries,left_on='Country or region',right_on='Name')
happiness_countries_merge
```

```
score_by_continent = happiness_countries_merge[['Continent','Score']].groupby('Continent').mean()
score_by_continent
```

```
# Matplotlib
plt.bar(score_by_continent.index, score_by_continent['Score'])
plt.xticks(rotation=90)
plt.show()
```

```
# Pandas
score_by_continent.plot(kind = 'bar')
```

```
# Seaborn
sns.barplot(x=score_by_continent.index, y=score_by_continent['Score'])
plt.xticks(rotation=90)
```
