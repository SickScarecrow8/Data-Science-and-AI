# **TEMA 2. Anotaciones**

1.	Crea un dataframe `cities` con la información del archivo world_table_city.csv

```
cities = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DS & AI - M4/M4 | Profundiza/world_table_city.csv")
cities
```

2.	A partir del dataframe `cities` obtén otro (`citiesMEX`) donde sólo almacenes las ciudades de México

```
citiesMEX = cities[(cities['CountryCode']=='MEX')]
citiesMEX
```

3.	Determinar el estado (`district`) con mayor población

```
citiesMEX = citiesMEX[['District','Population']].groupby('District').sum()
citiesMEX
```

```
maxDistrict = citiesMEX[citiesMEX['Population']==citiesMEX['Population'].max()]
maxDistrict
```

4.	Calcular el porcentaje que representa dicha cantidad en el total (suma de la población de todos los estados)

```
percentage = maxDistrict['Population'].to_list()[0]*100 / citiesMEX.sum()[0]
percentage
```

5.	Graficar el dataframe en un diagrama tipo pie

```
citiesMEX['Population'].plot(kind='pie',figsize=(10,10))
```

6.	Girar los nombres de los estados para que sean legibles en el diagrama

```
citiesMEX['Population'].plot(kind='pie',rotatelabels=True,figsize=(10,10))
```

7.	Colocar el porcentaje calculado como texto en la gráfica considerando coordenadas fijadas por el tamaño del gráfico (el punto central se encuentra en 0,0) y no por los datos.

```
citiesMEX['Population'].plot(kind='pie',rotatelabels=True,figsize=(10,10))
plt.text(-0.63,0.35,str(f'{percentage:.2f}')+'%',fontsize=20,fontweight='bold',color='white')
```
