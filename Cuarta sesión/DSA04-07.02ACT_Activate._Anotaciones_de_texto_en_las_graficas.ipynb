8. Ahora, crea un dataframe countries con la información del archivo world_table_country.csv

```
countries = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Science and AI/Cuarta sesión/world_table_country.csv')
countries
```

9. A partir del dataframe countries obtén otro (centralAmerica) como a continuación se indica para completar los siguientes pasos

```
centralAmerica = countries.loc[(countries['Continent'] == 'North America') & (countries['Region'] == 'Central America'),
                               ['Code', 'Name', 'SurfaceArea', 'Population', 'LifeExpectancy', 'GNP']]
centralAmerica.set_index('Code', inplace = True)
centralAmerica
```

10. Graficar todas sus columnas numéricas en un diagrama de barras

```

# Get the numerical columns you want to plot
numerical_cols = ['SurfaceArea', 'Population', 'LifeExpectancy', 'GNP']

# Set the width of the bars
bar_width = 0.2

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Iterate through the numerical columns and plot them as bars
for i, col in enumerate(numerical_cols):
    # Calculate the x positions of the bars for this column
    x_pos = range(len(centralAmerica.index))  # Original x positions
    x_pos = [x + i * bar_width for x in x_pos] # Shift x positions for each column

    # Plot the bars for the current column
    ax.bar(x_pos, centralAmerica[col], width=bar_width, label=col)

# Set x-axis ticks and labels
ax.set_xticks(range(len(centralAmerica.index)))
ax.set_xticklabels(centralAmerica.index)

# Customize the plot (add title, legend, labels, etc.)
ax.set_title('Central America Data')
ax.set_xlabel('Country Code')
ax.set_ylabel('Values')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
